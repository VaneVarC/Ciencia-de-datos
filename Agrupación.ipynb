#Algoritmos de agrupación en clústers en Python
El término análisis de clúster hace referencia a la familia de algoritmos que **permiten agrupar registros similares** de un conjunto de datos en grupos. El objetivo final del análisis es asignar a cada clúster los registros que son similares entre sí. 
- Técnica de aprendizaje no supervisado.

```python
#*******Importacion de librerias********
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generación de datos 
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.6, random_state=50)
```

```python
#Creación de array 
points = data[0]
```

```python
#Creación de grafico de dispersion
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.axhline(y=0,linestyle=":")  #Marca una lina para saber en donde esta el 0
plt.axvline(x=0,linestyle=":")
```

**Algoritmo K-Means**

```python
# importar liberia KMeans
from sklearn.cluster import KMeans
```

```python
# Crear kmeans object
kmeans = KMeans(n_clusters=4)
# fit kmeans object to data
kmeans.fit(points)
# print location of clusters learned by kmeans object
print(kmeans.cluster_centers_)
# save new clusters for chart
y_km = kmeans.fit_predict(points)
```

```python
#Graficas de dispersión asginado colores 
plt.axhline(y=0,linestyle=":")  #Marca una lina para saber en donde esta el 0
plt.axvline(x=0,linestyle=":")
plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='red')
plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='black')
plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='blue')
plt.scatter(points[y_km ==3,0], points[y_km == 3,1], s=100, c='cyan')
```

**Algoritmo hierarchy**

```python
# importar la libreria hierarchical cluster
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
# creación del cluster
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
# Guardar clusters para el grafico
y_hc = hc.fit_predict(points)
```

```python
#Graficas de dispersión asginado colores 
plt.axhline(y=0,linestyle=":")  #Marca una lina para saber en donde esta el 0
plt.axvline(x=0,linestyle=":")
plt.scatter(points[y_hc ==0,0], points[y_hc == 0,1], s=100, c='red')
plt.scatter(points[y_hc==1,0], points[y_hc == 1,1], s=100, c='black')
plt.scatter(points[y_hc ==2,0], points[y_hc == 2,1], s=100, c='blue')
plt.scatter(points[y_hc ==3,0], points[y_hc == 3,1], s=100, c='cyan')
```
